```python
import time_management as tm
import time

# Method 1  
# Specific Part of Code
with tm.block_timer("loop"): # add
    for i in range(2):
        time.sleep(0.1)

# Method 2  
# Function
@tm.func_timer # add
def my_function():
    for _ in range(2):
        time.sleep(0.1)

my_function()

# Method 3  
# Line by Line
def track_lines(): # add
    for _ in range(2):
        print('')

tm.line_timer(track_lines)
```
